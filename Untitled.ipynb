```python
from os import rename, listdir, makedirs
from os.path import isfile, join
import os
import random
from shutil import copyfile
import glob
import sys
import pandas as pd
import json
```

```python
all_images = [f for f in listdir("./all_images/") if isfile(join("./all_images/", f))]
```

```python
random24 = random.sample(all_images, 24)
```

```python
labels = pd.read_csv("pics3_labeling-100v2.csv")
```

```python
cols = list(labels)[2:]
```

```python
random24[1][0:len(random24[1])-9]
```

```python
labels["Name"][0][0:len(labels["Name"][0])-1]
```

```python
data = {}
for i in range(24):
    data["pic"+str(i)] = []
    index = -1
    count = 0
    for j in labels["Name"]:
        if random24[i][0:len(random24[i])-9]==j[0:len(j)-1]:
            index=count
        count+=1
    for c in cols:
        if (labels[c][index]=="Y"):
            data["pic"+str(i)].append({c:"yes"})
        else:
            data["pic"+str(i)].append({c:"no"})
```
